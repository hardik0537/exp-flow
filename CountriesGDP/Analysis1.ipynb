```python
# Reading data from csv file
countries_stats_df = pd.read_csv('countries.csv') 
# Removing labeled dimensions
countries_stats_df = countries_stats_df.drop('Country', axis=1)
countries_stats_df = countries_stats_df.drop('Region', axis=1)
# Removing NaN
countries_stats_df = countries_stats_df.fillna(0)
countries_stats_df = countries_stats_df.convert_objects(convert_numeric=True)

corr = countries_stats_df.corr(method='spearman')
sns.heatmap(corr)
```

The above graph uses spearman correlation to understand the dependency between the given features. The reason we have used Spearman correlation is because we have least ordinal feature values with no proof if they are all from the same distribution.

Based on above graph, the relationships are self explanatory with the given scale in the closed range [-1, +1].
If the square is towards blue shade, then it means both the variables are inversely relational to each other and direclty proportional if they are towards red.

Thus, to answer the question about a Country's GDP, we can say it is 
1. Directly related to Phones(per 1000), Literacy, Service, Net Immigration, and so on.
2. Inversely proportional to Infant mortality (per 1000 births), Birthrate, Agriculture, and so on.
